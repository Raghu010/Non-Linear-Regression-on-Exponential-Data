```python
x = np.arange(-5, 5, 0.1)
print(x.size)
y = 2*(x) + 3
y_noise = 2*np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()     # Fitting a Regression line in a scattered data
```

```python
x = np.arange(-5, 5, 0.1)
y = 1*(x**3) + 1*(x**2) + 1*(x) + 3 
y_noise = 20*np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()                      #polynomial Plot
```

```python
x = np.arange(-5, 5, 0.1)
y = np.power(x, 2)
y_noise = 2*np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.title('Quadratic or plob')
plt.show()                    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
import seaborn as sn
plt.figure(figsize = (10, 5))
x_data , y_data = (df['Year'].values, df['Value'].values)
plt.plot(x_data, y_data, 'bo')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = 1.0 / (1.0 + np.exp(-x))
plt.plot(x, y)
plt.ylabel('dependent variable')
plt.xlabel('independent variable')
plt.show()
```

```python
def sigmoid(x, Beta_1, Beta_2):
    y = 1/(1+np.exp(-Beta_1*(x-Beta_2)))
    return y
```

```python
Beta_1 = 0.10
Beta_2 = 1990.0
Y_pred = sigmoid(x_data, Beta_1, Beta_2)
plt.plot(x_data, Y_pred*150000000000000.)
plt.plot(x_data, y_data, 'ro')
```

```python
xdata = x_data/max(x_data)
print(xdata)
ydata = y_data/max(y_data)
ydata
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit (sigmoid, xdata, ydata)
print(popt)
print(pcov)
print('beta_1 = %f, beta_2 = %f' % (popt[0], popt[1]))
```

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize = (10, 5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'k*', label = 'data')
plt.plot(x, y, linewidth = 5.0, label = 'fit')
plt.legend(loc = 'best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]
popt, pcov = curve_fit(sigmoid, train_x, train_y)

y_hat = sigmoid(test_x, *popt)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score 
print(mean_absolute_error(y_hat,test_y))
print(r2_score(y_hat,test_y))
```
